```python

json_file_path = '/Users/koala/Documents/Lighthub/Compass/Taleau_Project/weekly_earnings.json'

# Read the JSON file
with open(json_file_path, 'r') as file:
    data_json = json.load(file)

# Extract the "data" section
data_section = data_json.get('data', [])
```

```python
weekly_earnings_df = pd.DataFrame(data_section,columns=df_columns)
```

```python
weekly_earnings_df.head()
```

```python
columns_to_drop=['sid','id','position','created_at','created_meta','updated_at','updated_meta','meta','NAICS']
weekly_earnings_df=weekly_earnings_df.drop(columns=columns_to_drop)
```

```python
column_order = ['When','Canada','Alberta','British Columbia','Manitoba','NewBrunswick','Newfoundland And Labrador','Nova Scotia','Ontario', 'Prince Edward Island','Quebec','Saskatchewan']
weekly_earnings_df=weekly_earnings_df[column_order]
```
